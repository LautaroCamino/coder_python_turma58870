```python
# Importar bibliotecas
import pandas as pd
import requests

# Requisi√ß√£o HTTP para a API, especificando a URL e os par√¢metros necess√°rios
url = 'https://restcountries.com/v3.1/all'
response = requests.get(url)

# Verificar se a requisi√ß√£o foi bem-sucedida, ou seja, se o c√≥digo de status HTTP retornado pela API √© 200.
# Adicionalmente, criar vari√°vel para depois criar df
if response.status_code == 200:
    data_json = response.json()
else:
    print ('Erro ao acessar a API')

#Verificar estrutura
data_json[1]
```
