```python
import sqlite3
import pandas as pd
```

```python
def tabelas_bd():
    '''
       Retorna um dataframe com as tabelas do banco de dados 
    '''
#Conectar ao banco de dados
    conn = sqlite3.connect('coderhouse.db')
#Executar consulta que retorna as informações do esquema do banco de dados
    query = "SELECT name FROM sqlite_master WHERE type='table'"
    schema = pd.read_sql_query(query,conn)
#Fechar conexão
    conn.close()
    return schema
```

```python
def salva_bd(df, nome_tabela):
    '''
        Salva dataframe df na tabela nome_tabela.
    '''
#Conectar ao banco de dados
    conn = sqlite3.connect('coderhouse.db')
# Salvar o dataframe df na tabela nome_tabela. Se uma tabela com o mesmo nome já existir, ela será substituída
    df.to_sql(nome_tabela, conn, if_exists='replace', index = False)
    conn.close()
    return True
```

```python
def carrega_bd(nome_tabela):
    '''
        Carrega tabela nome_tabela num dataframe.
    '''
#Conectar ao banco de dados
    conn = sqlite3.connect('coderhouse.db')
#Executar a consulta SQL e retornar os resultados como um dataframe
#Selecionar todos os dados da tabela nome_tabela
    df = pd.read_sql_query(f'SELECT * FROM {nome_tabela}', conn)
#Fechar conexao
    conn.close()
    return df
```

```python
#Abrir arquivo xlsx
excel = pd.read_excel('fastfood.xlsx')
excel.head(3)
```

```python
tabelas_bd()
```

```python
salva_bd(excel,'fastfood')
```

```python
carrega_bd('fastfood')
```
